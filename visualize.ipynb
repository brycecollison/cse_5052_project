```python
import scipy.io
import numpy as np
from matplotlib import pyplot as plt

data = scipy.io.loadmat('mnist_49_3000.mat')
x = np.array(data['x'])
y = np.array(data['y'][0])
y[y==-1] = 0
#To visualize an image, type the following:
index = 0 #change the index to show different images
print(y[index])
image = x[:,index].reshape(28,28)
plt.imshow(image, interpolation='nearest')
plt.show()
plt.close()
```

```python
fig = plt.figure(constrained_layout=True)
fig.suptitle('Classified Images')

subfigs = fig.subfigures(nrows=3, ncols=1)

for row, subfig in enumerate(subfigs):
  subfig.suptitle(f'Row {row}')

  axs = subfig.subplots(nrows=1, ncols=5)
  for col, ax in enumerate(axs):
    ax.imshow(x[:,col].reshape(28,28))
    ax.axis('off')
    ax.set_title(f'Plot title {col}')
```
